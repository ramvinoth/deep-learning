## Visualization of the data

Not all the features in the dataset are useful for the purpose of regression. We select only perticular features which can be actually used further in the program for plotting.

```python
plt.rcParams["figure.figsize"] = [6,3]
df.hist(column='Open')
df.hist(column='High')
df.hist(column='Low')
df.hist(column='Close')
plt.tight_layout()
```

#### Observation:
We can clearly see the correlation between the Open, High, Low, Close data. This shows that the price range is not too dynamic within short period of time.

```python
plt.rcParams["figure.figsize"] = [15,8]

X = df['High'].values
#X = X.reshape(-1,len(X))
#print(X)
Y = df['Low'].values
#Y = Y.reshape(-1,len(Y))
#print(Y)

plt.xlabel('Low Value')
plt.ylabel('High Value')

print(plt.scatter(X,Y))
```

#### Observation:

#### Stochastic Oscillator

Using %K and %D, we can build a stochastic oscilator which is one of the famous technical indicatiors in market analysis, to show the change of trend in price.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20,10))
df['Close'][:100].plot(ax=axes[0])#, title='Close'
df[['%K', '%D']][:1000].plot(ax=axes[1])#, title='Oscillator'
plt.show()
```

#### Observation:

From the above chart, we can infer the change in price movement direction whenever the stochastic oscilator goes extreme top(100) or down(0)
